### For Category and Attribute

```python
cat_directory = pwd + '/drive/MyDrive/DeepFashion'+category_path
data_path = cat_directory + '/Img/'
bbox_path = cat_directory + '/Anno_coarse/list_bbox.txt'
img_category_path = cat_directory + '/Anno_coarse/list_category_img.txt'
```

```python
img_bbox = {}

try:
    bbox_file = open(bbox_path, 'r')
except Exception:
    print('get_bbox() error')

bbox_data = bbox_file.readlines()
for d, i in tqdm(zip(bbox_data[2:],range(bbox_data.__len__()-2))):
    s = d.split()
    p = data_path + s[0]
    bbox_loca = []
    for a in s[1:]:
        bbox_loca.append(int(a))
    img_bbox[p] = bbox_loca
bbox_file.close()
```

```python
img_category = {}

#count_1 = 0
#count_2 = 0
#count_3 = 0

try:
    img_category_file = open(img_category_path, 'r')
except Exception:
    print ('img_category_path error!!')
lines = img_category_file.readlines()
for l in tqdm(lines[2:]):
    s = l.split()
    path = data_path + s[0]

    s[1] = int(s[1])

    # 1~20 is (1), 21~36 is (2), 37~50 is (3)
    if s[1] < 21 :
        img_category[path] = 1
        #count_1 += 1
    elif s[1] > 36 :
        img_category[path] = 3
        #count_3 += 1
    else :
        img_category[path] = 2
        #count_2 += 1

#print count_1
#print count_2
#print count_3
#print count_1 + count_2 + count_3
```

```python
img_bbox
```

```python
# merge dictionaries together

ds = [img_bbox, img_category]
d = {}
for k in tqdm(img_bbox.keys()):
    d[k] = tuple(d[k] for d in ds)
```

```python
line = ""
f = open(output_path + 'd-category-all.txt', 'w')
for key,value in tqdm(d.items()):
    line = str(key)
    for v in value:
        if type(v) == list:
            for point in v:
                line += ": " + str(point)
        else:
            line += ": " + str(v) + '\n'

    if int(line.split(":")[-1]) < 3 :
        f.write(line)

bbox_file.close()
f.close()
```

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train_path = pwd + '/Data/Train/'
test_path = pwd + '/Data/Test/'
if not os.path.exists(train_path):
    os.makedirs(train_path)
if not os.path.exists(test_path):
    os.makedirs(test_path)
```

```python
all_files = ['d-category-all.txt', 'd-consumer-all.txt',
            'd-inshop-all.txt', 'd-landmark-all.txt']

all_files2 = ['d-category-all.txt', 'd-landmark-all.txt']

TEST_SIZE = 0.2
```

```python
# verify length of each files
for fname in all_files:
    with open(output_path + fname) as f:
        print(fname, len(f.readlines()))

# total-category : 28922
# total-consumer : 239557
# total-inshop : 52712
# total-landmark : 123016

# discrepancy in total because "3" (full-body) is not included in .txt files
```

```python
# split train/test set
for fname in all_files:
    with open(output_path + fname) as f:
        data = f.read().split('\n')
        data = np.array(data)

        x_train, x_test = train_test_split(data, test_size = TEST_SIZE)

        print(fname)
        print("train: ", len(x_train))
        print("test: ", len(x_test))
        print("total: ", len(x_train)+len(x_test))
        print("")

        s, _ =  fname.split('all')

        with open(train_path+s+'train.txt','w') as f_train:
            for item in x_train:
                if len(item) > 1:
                    f_train.write(item + '\n')
        with open(test_path+s+'test.txt','w') as f_test:
            for item in x_test:
                if len(item) > 1:
                    f_test.write(item + '\n')
```
